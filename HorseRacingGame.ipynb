```python
import random
import time

# =========================== Horse Data ===========================
horses = [
    {"name": "Gold Ship", "wins": 0},
    {"name": "Haru Urara", "wins": 0},
    {"name": "Bakushin O", "wins": 0},
    {"name": "Tokai Teio", "wins": 0},
    {"name": "Oguri Cap", "wins": 0},
    {"name": "Rice Shower", "wins": 0},
    {"name": "Special Week", "wins": 0},
    {"name": "Lionel Messi", "wins": 0},
    {"name": "Khan Kluay", "wins": 0},
    {"name": "Verstappen", "wins": 0},
]

# =========================== Tie-Breaker ===========================
def tie_breaker(tied_horses):
    finish_line = 30
    positions = [0] * len(tied_horses)

    print("\n🏁 Final Stretch! 🏁\n")

    while max(positions) < finish_line:
        print("📸 SLOW MOTION 📸\n")
        for i, horse in enumerate(tied_horses):
            move = random.randint(1, 5)
            positions[i] = min(positions[i] + move, finish_line)
            track = "-" * positions[i]
            spaces = " " * (finish_line - positions[i])
            print(f"{horse['name']:<12}: {track}🐎>{spaces}|")
        time.sleep(0.8)

    # Decide winner
    max_pos = max(positions)
    winners = [tied_horses[i] for i, pos in enumerate(positions) if pos == max_pos]

    if len(winners) == 1:
        winners[0]["wins"] += 1
        print(f"\n🏆 Winner: {winners[0]['name']}! 🏆\n")
    else:
        tied_names = ", ".join([w['name'] for w in winners])
        print(f"\n🔥 DEAD HEAT! Co-winners: {tied_names}! 🏆\n")
        for w in winners:
            w["wins"] += 1

# =========================== Race Simulation ===========================
def run_race(horses):
    finish_line = 50
    positions = [0] * len(horses)

    while max(positions) < finish_line:
        print("🚩 Horse Race Simulation 🚩\n")
        for i, horse in enumerate(horses):
            move = random.randint(1, 3)
            positions[i] = min(positions[i] + move, finish_line)
            track = "-" * positions[i]
            spaces = " " * (finish_line - positions[i])
            print(f"{horse['name']:<12}: {track}🐎>{spaces}|")
        time.sleep(0.3)

    # Determine winner(s)
    max_pos = max(positions)
    winners = [horses[i] for i, pos in enumerate(positions) if pos == max_pos]

    if len(winners) == 1:
        winners[0]["wins"] += 1
        print(f"\n🏆 Winner: {winners[0]['name']}! 🏆\n")
    else:
        tied_names = ", ".join([w["name"] for w in winners])
        print(f"\n😱 They're so close! 😱\n")
        tie_breaker(winners)

# =========================== Sorting ===========================
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L, R = arr[:mid], arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i]["wins"] <= R[j]["wins"]:
                arr[k] = L[i]; i += 1
            else:
                arr[k] = R[j]; j += 1
            k += 1

        while i < len(L): arr[k] = L[i]; i += 1; k += 1
        while j < len(R): arr[k] = R[j]; j += 1; k += 1
    return arr

def sort_by_wins():
    horses_copy = horses.copy()
    merge_sort(horses_copy)
    horses_copy.reverse()
    top_five = horses_copy[:5]

    print("\n📊 Top 5 Leaderboard (by wins):")
    for i, horse in enumerate(top_five, start=1):
        print(f"{i}. {horse['name']} - {horse['wins']} wins")

# =========================== Searching ===========================
def search_horse(name):
    for horse in horses:
        if horse["name"].lower() == name.lower():
            print(f"🔎 Found {horse['name']} - Wins: {horse['wins']}")
            return
    print("\n⚠️ Horse not found.")

# =========================== Insertion ===========================
def add_horse(name):
    if any(h["name"].lower() == name.lower() for h in horses):
      print("\n❌ Horse with that name already exists!")
    else:
      horses.append({"name": name, "wins": 0})
      print(f"\n✅ Horse '{name}' has been added to the race lineup!")

# =========================== Deletion ===========================
def remove_horse(name):
    global horses
    before = len(horses)
    horses = [h for h in horses if h["name"].lower() != name.lower()]
    if len(horses) < before:
        print(f"💔 Horse '{name}' removed from the race lineup.")
    else:
        print("\n⚠️ Horse not found.")

# =========================== All horses ===========================
def all_horse():
    print("\n🐎 Current Horses:")
    for h in horses:
        print(f"- {h['name']} (Wins: {h['wins']})")

# =========================== Menu System ===========================
def main_menu():
    print(r"""
  ██     ██   ██  ██████  ██████  ███████ ███████     ██████   █████   ██████ ██ ███    ██  ██████       ██████   █████  ███    ███ ███████ ██
 ██      ██   ██ ██    ██ ██   ██ ██      ██          ██   ██ ██   ██ ██      ██ ████   ██ ██           ██       ██   ██ ████  ████ ██       ██
██       ███████ ██    ██ ██████  ███████ █████       ██████  ███████ ██      ██ ██ ██  ██ ██   ███     ██   ███ ███████ ██ ████ ██ █████     ██
 ██      ██   ██ ██    ██ ██   ██      ██ ██          ██   ██ ██   ██ ██      ██ ██  ██ ██ ██    ██     ██    ██ ██   ██ ██  ██  ██ ██       ██
  ██     ██   ██  ██████  ██   ██ ███████ ███████     ██   ██ ██   ██  ██████ ██ ██   ████  ██████       ██████  ██   ██ ██      ██ ███████ ██
    """)
    while True:
        print("\n=== Main Menu ===")
        print("1. Run a race")
        print("2. Show leaderboard (sorted by wins)")
        print("3. Search for a horse")
        print("4. Add a new horse")
        print("5. Remove a horse")
        print("6. Show all horses")
        print("0. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            if len(horses) == 0:
              print("\n❌There are no horses!")
            elif len(horses) == 1:
              print("\n🌹You can't race by yourself...")
            else:
              run_race(horses)
        elif choice == "2":
            if len(horses) == 0:
              print("\n❌There are no horses!")
            else:
              sort_by_wins()
        elif choice == "3":
            search_horse(input("Enter horse name: "))
        elif choice == "4":
            add_horse(input("Enter new horse name: "))
        elif choice == "5":
            if len(horses) == 0:
              print("\n❌There are no horses!")
            else:
              remove_horse(input("Enter horse name to remove: "))
        elif choice == "6":
            all_horse()
        elif choice == "0":
            print("Exiting game...")
            print(r"""
 ██████   ██████   ██████  ██████      ██████  ██    ██ ███████ ██
██       ██    ██ ██    ██ ██   ██     ██   ██  ██  ██  ██      ██
██   ███ ██    ██ ██    ██ ██   ██     ██████    ████   █████   ██
██    ██ ██    ██ ██    ██ ██   ██     ██   ██    ██    ██
 ██████   ██████   ██████  ██████      ██████     ██    ███████ ██
    """)
            break
        else:
            print("⚠️ Invalid choice. Try again.")
```
